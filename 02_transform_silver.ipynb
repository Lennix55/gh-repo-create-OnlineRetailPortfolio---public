```python
df = spark.table("bronze_online_retail")

display(df)
```

```python
df.printSchema()
```

```python
from pyspark.sql.functions import col

clean_df = (
    df
    .withColumn("UnitPrice", col("Price").cast("float"))
    .withColumn("CustomerID", col("CustomerID").cast("string"))
)
```

```python
clean_df = clean_df.dropna(subset=["Invoice"])
```

```python
clean_df = clean_df.dropDuplicates()
```

```python
clean_df = clean_df.withColumn(
    "LineTotal",
    col("Quantity") * col("UnitPrice")
)
```

```python
clean_df.write.mode("overwrite").saveAsTable("silver_online_retail")
```
