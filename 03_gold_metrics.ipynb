```python
df = spark.table("silver_online_retail")
display(df)
```

```python
%%sql
CREATE OR REPLACE TABLE gold_daily_sales AS
SELECT
    TO_DATE(InvoiceDate, 'MM-dd-yyyy HH:mm') AS OrderDate,
    SUM(Quantity * UnitPrice) AS RevenueThatDay,
    COUNT(DISTINCT Invoice) AS TotalTransactionsThatDay
FROM silver_online_retail
GROUP BY TO_DATE(InvoiceDate, 'MM-dd-yyyy HH:mm')
ORDER BY OrderDate DESC;

```

```python
customer_sales = (
    df.groupBy("CustomerID")
    .agg(
        sum("LineTotal").alias("CustomerLifetimeValue"),
        count("Invoice").alias("TotalOrders")
    )
)

customer_sales.write.mode("overwrite").saveAsTable("gold_customer_metrics")
```

```python
product_sales = (
    df.groupBy("StockCode", "Description")
    .agg(
        sum("Quantity").alias("UnitsSold"),
        sum("LineTotal").alias("Revenue")
    )
)

product_sales.write.mode("overwrite").saveAsTable("gold_product_metrics")
```
